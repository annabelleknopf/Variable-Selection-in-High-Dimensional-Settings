
# Prep_Results_Sim2 — GWAS-like Simulation Analysis


```python
#imports
import os
import glob
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import math
import textwrap
import warnings
warnings.filterwarnings("ignore", message=".*observed=False is deprecated.*")
pd.options.display.float_format = "{:.2f}".format
```

```python
RESULTS_DIR = "Results2"      
PREP_DIR = os.path.join(RESULTS_DIR, "Prepared2")
os.makedirs(PREP_DIR, exist_ok=True)
print("Nutze Ergebnisordner:", os.path.abspath(PREP_DIR))

# ---------------------------------------------
# Loader
# ---------------------------------------------
def load_all_pickles(results_dir: str):
    """
    Lädt alle Pickle-Dateien der Sim2-Serie:
      - Sim2T0.k_0_<p>.pkl         (null)
      - Sim2a.k_<k>_<p>.pkl        (gamma)
      - Sim2b.k_<k>_<p>.pkl        (normal)
    und gibt eine Liste von Dicts zurück.
    """
    pattern = os.path.join(results_dir, "Sim2*.k_*.pkl")
    files = sorted(glob.glob(pattern))
    if not files:
        print("⚠️ Keine Dateien gefunden unter", pattern)
    items = []
    for fn in files:
        try:
            with open(fn, "rb") as f:
                data = pickle.load(f)
            # Fallbacks & Konsistenz
            base = os.path.basename(fn)
            data["__file__"] = fn
            # k, p, scenario sind (laut Sim2) bereits im Pickle enthalten – sonst konservativ aus Dateinamen schätzen
            if "p" not in data:
                try:
                    # ... letzte Komponente nach dem letzten Unterstrich vor .pkl
                    p_str = base.split(".k_")[1].split(".pkl")[0].split("_")[-1]
                    data["p"] = int(p_str)
                except Exception:
                    pass
            if "k" not in data:
                try:
                    k_str = base.split(".k_")[1].split("_")[0]
                    data["k"] = int(k_str)
                except Exception:
                    pass
            if "scenario" not in data:
                # aus Dateinamen heuristisch: T0 → null, a → gamma, b → normal
                if "Sim2T0" in base:
                    data["scenario"] = "null"
                elif "Sim2a" in base:
                    data["scenario"] = "gamma"
                elif "Sim2b" in base:
                    data["scenario"] = "normal"
            items.append(data)
        except Exception as e:
            print(f"Fehler beim Laden von {fn}: {e}")
    return items

# ---------------------------------------------
# Helper: Matrix → Long-Format
# ---------------------------------------------
def _melt_matrix(mat: np.ndarray,
                 method_names: list[str],
                 sim_offset: int,
                 p: int,
                 k: int,
                 scenario: str,
                 metric: str,
                 criterion: str | None,
                 source_file: str) -> pd.DataFrame:
    """Formt (sim_nr × nr_methods)-Matrix in Long-Format um."""
    sim_nr, nr_methods = mat.shape
    assert nr_methods == len(method_names), "method_names passt nicht zur Matrixbreite."

    df = pd.DataFrame(mat, columns=method_names)
    df["sim"] = np.arange(sim_nr) + sim_offset
    long_df = df.melt(id_vars=["sim"], var_name="method", value_name="value")
    long_df["metric"] = metric                # "score", "FP", "TP", "runtime"
    long_df["criterion"] = criterion          # "mBIC", "mBIC2" oder None
    long_df["p"] = p
    long_df["k"] = k
    long_df["scenario"] = scenario            # "null" | "gamma" | "normal"
    long_df["source_file"] = source_file
    return long_df

# ---------------------------------------------
# Hauptfunktion: baut ein gesamtes Long-DF
# ---------------------------------------------
def build_sim2_long_df(results_dir: str = RESULTS_DIR) -> pd.DataFrame:
    items = load_all_pickles(results_dir)
    out_frames = []
    sim_offset = 0

    for it in items:
        try:
            m1 = it["mBIC.results"]     # (sim × methods)
            m2 = it["mBIC2.results"]
            fp1 = it["mBIC.FP"]
            fp2 = it["mBIC2.FP"]
            tp1 = it["mBIC.TP"]
            tp2 = it["mBIC2.TP"]
            rt = it["runtime"]
            method_names = list(it["method_names"])
            p = int(it.get("p"))
            k = int(it.get("k"))
            scenario = str(it.get("scenario"))
            src = it["__file__"]

            # sim_nr für den Offset
            sim_nr, nr_methods = m1.shape

            # mBIC / mBIC2 Scores
            out_frames.append(_melt_matrix(m1, method_names, sim_offset, p, k, scenario, "score", "mBIC",  src))
            out_frames.append(_melt_matrix(m2, method_names, sim_offset, p, k, scenario, "score", "mBIC2", src))

            # False Positives / True Positives (je Kriterium)
            out_frames.append(_melt_matrix(fp1, method_names, sim_offset, p, k, scenario, "FP", "mBIC",  src))
            out_frames.append(_melt_matrix(fp2, method_names, sim_offset, p, k, scenario, "FP", "mBIC2", src))
            out_frames.append(_melt_matrix(tp1, method_names, sim_offset, p, k, scenario, "TP", "mBIC",  src))
            out_frames.append(_melt_matrix(tp2, method_names, sim_offset, p, k, scenario, "TP", "mBIC2", src))

            # Runtime (kein Kriterium)
            out_frames.append(_melt_matrix(rt, method_names, sim_offset, p, k, scenario, "runtime", None, src))

            # Nächster Block bekommt fortlaufende sim-IDs
            sim_offset += sim_nr

        except KeyError as e:
            print(f"⚠️ Schlüssel fehlt in {it.get('__file__', '?')}: {e}")

    if not out_frames:
        return pd.DataFrame(columns=["sim","method","metric","criterion","value","p","k","scenario","source_file"])

    df_long = pd.concat(out_frames, ignore_index=True)

    # Speicher optional eine vorbereitete Kopie
    out_fn = os.path.join(PREP_DIR, "Sim2_long.parquet")
    try:
        df_long.to_parquet(out_fn, index=False)
        print("Gespeichert:", out_fn)
    except Exception as e:
        print("Hinweis: Parquet-Speicherung schlug fehl, ggf. pyarrow/fastparquet installieren. Fehler:", e)

    return df_long
```

```python
EPS = globals().get("EPS", 1e-8)
EXCLUDE_METHODS = {"stepwise_ff"}  

# ---------------------------------------------
# Helper
# ---------------------------------------------
def baseline_diff_by_name(M, method_names, baseline="stepwise_plain"):
    """
    Subtrahiert die Spalte der Baseline-Methode (falls vorhanden),
    sonst fällt auf Spalte 0 zurück. Liefert (Diff, baseline_index).
    """
    if baseline in method_names:
        j = method_names.index(baseline)
    else:
        j = 0
    base = M[:, [j]]
    return M - base, j

def as_df_long_diff_sim2(diff_mat, method_names, instance_offset, scenario_label, value_name, k, p):
    sim_nr, nr_methods = diff_mat.shape
    df = pd.DataFrame(diff_mat, columns=method_names)
    df["instance"] = np.arange(sim_nr) + instance_offset
    df["scenario"] = scenario_label  # "null" | "gamma" | "normal"
    long_df = df.melt(id_vars=["instance", "scenario"], var_name="variable", value_name=value_name)
    long_df["k"] = k
    long_df["p"] = p
    return long_df

# ---------------------------------------------
# Laden & Methodenreihenfolge 
# ---------------------------------------------
all_data = load_all_pickles(RESULTS_DIR)

method_order = []
for blk in all_data:
    for m in blk["method_names"]:
        if m in EXCLUDE_METHODS:
            continue
        if m not in method_order:
            method_order.append(m)

# ---------------------------------------------
# Aggregation (ohne stepwise_ff)
# ---------------------------------------------
df_rows = []
df_mBIC_list = []
df_mBIC2_list = []
df_Runtime_list = []

instance_offset = 0  # fortlaufende Sim-IDs

for block in all_data:
    # Rohmatrizen laden
    mBIC_results  = np.asarray(block["mBIC.results"])
    mBIC2_results = np.asarray(block["mBIC2.results"])
    mBIC_FP       = np.asarray(block["mBIC.FP"])
    mBIC2_FP      = np.asarray(block["mBIC2.FP"])
    mBIC_TP       = np.asarray(block["mBIC.TP"])
    mBIC2_TP      = np.asarray(block["mBIC2.TP"])
    runtime       = np.asarray(block["runtime"])
    method_names0 = list(block["method_names"])
    k             = int(block["k"])
    p             = int(block["p"])
    scenario_lab  = str(block["scenario"])

    # Spaltenmaske: stepwise_ff entfernen
    keep_mask = np.array([m not in EXCLUDE_METHODS for m in method_names0])
    if not keep_mask.any():
        # Falls ein Block nur stepwise_ff enthalten sollte, überspringen
        continue

    # Filtern
    mBIC_results  = mBIC_results[:,  keep_mask]
    mBIC2_results = mBIC2_results[:, keep_mask]
    mBIC_FP       = mBIC_FP[:,       keep_mask]
    mBIC2_FP      = mBIC2_FP[:,      keep_mask]
    mBIC_TP       = mBIC_TP[:,       keep_mask]
    mBIC2_TP      = mBIC2_TP[:,      keep_mask]
    runtime       = runtime[:,       keep_mask]
    method_names  = [m for m, keep in zip(method_names0, keep_mask) if keep]

    sim_nr, nr_methods = mBIC_results.shape

    # Baseline-Differenzen robust (bevorzugt stepwise_plain, sonst erste Spalte)
    res,  j1 = baseline_diff_by_name(mBIC_results,  method_names, baseline="stepwise_plain")
    res2, j2 = baseline_diff_by_name(mBIC2_results, method_names, baseline="stepwise_plain")

    # NaN-robust
    valid_res  = ~np.isnan(res)
    valid_res2 = ~np.isnan(res2)

    Worse   = np.sum((res  >  EPS) & valid_res,  axis=0).astype(int)
    Better  = np.sum((res  < -EPS) & valid_res,  axis=0).astype(int)
    Worse2  = np.sum((res2 >  EPS) & valid_res2, axis=0).astype(int)
    Better2 = np.sum((res2 < -EPS) & valid_res2, axis=0).astype(int)

    # FP/TP/Power
    FP = np.nanmean(mBIC_FP, axis=0)
    TP = np.nanmean(mBIC_TP, axis=0)
    Power = (TP / k) if k > 0 else np.zeros_like(TP)

    FP2 = np.nanmean(mBIC2_FP, axis=0)
    TP2 = np.nanmean(mBIC2_TP, axis=0)
    Power2 = (TP2 / k) if k > 0 else np.zeros_like(TP2)

    # FDR2 sim-weise
    denom = mBIC2_FP + mBIC2_TP
    with np.errstate(divide="ignore", invalid="ignore"):
        fdr2_sim = np.where(denom > 0, mBIC2_FP / denom, 0.0)
    FDR2 = np.nanmean(fdr2_sim, axis=0)

    # Mittel/SD/SEM der Differenzen
    n_eff_mBIC  = np.sum(valid_res,  axis=0)
    n_eff_mBIC2 = np.sum(valid_res2, axis=0)

    mean_mBIC = np.nanmean(res,  axis=0)
    sd_mBIC   = np.nanstd(res,   axis=0, ddof=1)
    sem_mBIC  = np.where(n_eff_mBIC  > 0, sd_mBIC  / np.sqrt(n_eff_mBIC),  np.nan)

    mean_mBIC2 = np.nanmean(res2, axis=0)
    sd_mBIC2   = np.nanstd(res2,  axis=0, ddof=1)
    sem_mBIC2  = np.where(n_eff_mBIC2 > 0, sd_mBIC2 / np.sqrt(n_eff_mBIC2), np.nan)

    # Runtime
    Runtime = np.nanmean(runtime, axis=0)

    # Aggregation pro Methode
    for l, mname in enumerate(method_names):
        df_rows.append({
            "k": k, "p": p, "scenario": scenario_lab, "method": mname,
            "Better": int(Better[l]), "Worse": int(Worse[l]),
            "Power": float(Power[l]), "FP": float(FP[l]),
            "Better2": int(Better2[l]), "Worse2": int(Worse2[l]),
            "Power2": float(Power2[l]), "FDR2": float(FDR2[l]),
            "mean_mBIC": float(mean_mBIC[l]), "sd_mBIC": float(sd_mBIC[l]), "sem_mBIC": float(sem_mBIC[l]),
            "mean_mBIC2": float(mean_mBIC2[l]), "sd_mBIC2": float(sd_mBIC2[l]), "sem_mBIC2": float(sem_mBIC2[l]),
            "Runtime": float(Runtime[l]),
        })

    # Long-Form
    df_mBIC_list.append(as_df_long_diff_sim2(res,  method_names, instance_offset, scenario_lab, "mBIC",  k, p))
    df_mBIC2_list.append(as_df_long_diff_sim2(res2, method_names, instance_offset, scenario_lab, "mBIC2", k, p))

    # Runtime Long-Form
    rt_df = pd.DataFrame(runtime, columns=method_names)
    rt_df["instance"] = np.arange(sim_nr) + instance_offset
    rt_df["scenario"] = scenario_lab
    runtime_long = rt_df.melt(id_vars=["instance", "scenario"], var_name="variable", value_name="Runtime")
    runtime_long["k"] = k
    runtime_long["p"] = p
    df_Runtime_list.append(runtime_long)

    # fortlaufende Sim-IDs
    instance_offset += sim_nr

# ---------------------------------------------
# Zusammenbauen & Kategorien
# ---------------------------------------------
df = pd.DataFrame(df_rows)
df_mBIC    = pd.concat(df_mBIC_list,  ignore_index=True) if df_mBIC_list  else pd.DataFrame(columns=["instance","scenario","variable","mBIC","k","p"])
df_mBIC2   = pd.concat(df_mBIC2_list, ignore_index=True) if df_mBIC2_list else pd.DataFrame(columns=["instance","scenario","variable","mBIC2","k","p"])
df_Runtime = pd.concat(df_Runtime_list,ignore_index=True) if df_Runtime_list else pd.DataFrame(columns=["instance","scenario","variable","Runtime","k","p"])

# Methoden-Reihenfolge setzen (ohne stepwise_ff)
if not df.empty:
    df["method"] = pd.Categorical(df["method"], categories=method_order, ordered=True)

for d in (df_mBIC, df_mBIC2, df_Runtime):
    if not d.empty:
        d["variable"] = pd.Categorical(d["variable"], categories=method_order, ordered=True)
        for col in ["instance", "scenario"]:
            if col in d.columns:
                d[col] = d[col].astype("category")

print("Fertig geladen . df shape:", df.shape)
```

```python
# --- Pivot Helper: wie dcast, aber für Sim2 mit (k, p, scenario) ---
def dcast_like_sim2(data, value_col, drop_methods=None, method_order=None):
    temp = data.copy()
    if drop_methods is not None and "method" in temp.columns:
        temp = temp[~temp["method"].isin(drop_methods)]

    index_cols = ["k", "p", "scenario"]
    pt = temp.pivot_table(index=index_cols, columns="method",
                          values=value_col, aggfunc="first").reset_index()
    pt.columns.name = None  # "method" Spaltenlabel entfernen

    # Spalten neu ordnen: erst Index, dann Methoden in method_order (falls vorhanden)
    method_cols = [c for c in pt.columns if c not in index_cols]
    if method_order is not None:
        ordered = [m for m in method_order if m in method_cols]
        leftovers = [m for m in method_cols if m not in ordered]
        cols = index_cols + ordered + leftovers
    else:
        cols = index_cols + method_cols
    return pt[cols]

# --- Build tables ---
drop_stepwise = ["stepwise_plain"]  # Baseline aus den Vergleichstabellen entfernen (optional)

tab_Better   = dcast_like_sim2(df, "Better",  drop_methods=drop_stepwise, method_order=method_order)
tab_Better2  = dcast_like_sim2(df, "Better2", drop_methods=drop_stepwise, method_order=method_order)
tab_Worse    = dcast_like_sim2(df, "Worse",   drop_methods=drop_stepwise, method_order=method_order)
tab_Worse2   = dcast_like_sim2(df, "Worse2",  drop_methods=drop_stepwise, method_order=method_order)
tabFP        = dcast_like_sim2(df, "FP",      method_order=method_order)
tabFDR2      = dcast_like_sim2(df, "FDR2",    method_order=method_order)

# Power-Tabellen nur für k > 0
df_posk   = df[df["k"] > 0] if not df.empty else df
tabPower  = dcast_like_sim2(df_posk, "Power",  method_order=method_order)
tabPower2 = dcast_like_sim2(df_posk, "Power2", method_order=method_order)

tab_m       = dcast_like_sim2(df, "mean_mBIC",  method_order=method_order)
tabsem      = dcast_like_sim2(df, "sem_mBIC",   method_order=method_order)
tab_m2      = dcast_like_sim2(df, "mean_mBIC2", method_order=method_order)
tabsem2     = dcast_like_sim2(df, "sem_mBIC2",  method_order=method_order)
tab_runtime = dcast_like_sim2(df, "Runtime",    method_order=method_order)

# --- Abspeichern ---
# Erwartet, dass PREP_DIR existiert (wie zuvor eingerichtet):
os.makedirs(PREP_DIR, exist_ok=True)

tables = {
    "Better":   tab_Better,
    "Better2":  tab_Better2,
    "Worse":    tab_Worse,
    "Worse2":   tab_Worse2,
    "FP":       tabFP,
    "FDR2":     tabFDR2,
    "Power":    tabPower,
    "Power2":   tabPower2,
    "mean_mBIC":  tab_m,
    "sem_mBIC":   tabsem,
    "mean_mBIC2": tab_m2,
    "sem_mBIC2":  tabsem2,
    "Runtime":    tab_runtime,
}

# 1) CSVs
for name, tbl in tables.items():
    out_csv = os.path.join(PREP_DIR, f"Sim2_{name}.csv")
    tbl.to_csv(out_csv, index=False)

print("Tabellen gespeichert unter:", os.path.abspath(PREP_DIR))
tab_Better2
```

# Results stratified by factors

```python
SC_MAP = {"gamma": "a", "normal": "b"}  # Szenario-Kürzel
K_LEVELS = [10, 20]
P_LEVELS = [1000, 2000, 5000, 10000]

os.makedirs(PREP_DIR, exist_ok=True)

def _reindex_methods_left(pt: pd.DataFrame) -> pd.DataFrame:
    pt = pt.reindex(method_order)
    pt.index.name = "method"
    pt.columns.name = None     # << Fix: keine „k“ / „p“ / „scenario“-Headerzelle
    return pt.reset_index()

def _tbl_by_k(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["k"].isin([10, 20]) & df_in["scenario"].isin(["gamma", "normal"])]
    g = d.groupby(["method", "k"])[value_col].sum().unstack("k", fill_value=0)
    for k in [10, 20]:
        if k not in g.columns:
            g[k] = 0
    g["all"] = g[[10, 20]].sum(axis=1)
    g = g.rename(columns={10: "k=10", 20: "k=20"})
    g.columns.name = None      # << Fix hier zusätzlich
    return _reindex_methods_left(g)

def _tbl_by_p(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["scenario"].isin(["gamma", "normal"]) & (df_in["k"] > 0) & df_in["p"].isin([1000,2000,5000,10000])]
    g = d.groupby(["method", "p"])[value_col].sum().unstack("p", fill_value=0)
    for pp in [1000,2000,5000,10000]:
        if pp not in g.columns: g[pp] = 0
    g = g[[1000,2000,5000,10000]]
    g.columns = [f"p={pp}" for pp in g.columns]
    g.columns.name = None      # << Fix
    return _reindex_methods_left(g)

def _tbl_by_scenario(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["scenario"].isin(["gamma", "normal"]) & (df_in["k"] > 0)]
    g = d.groupby(["method", "scenario"])[value_col].sum().unstack("scenario", fill_value=0)
    g = g.rename(columns={"gamma": "a", "normal": "b"})
    for c in ["a","b"]:
        if c not in g.columns: g[c] = 0
    g["all"] = g[["a","b"]].sum(axis=1)
    g = g[["a","b","all"]]
    g.columns.name = None      # << Fix
    return _reindex_methods_left(g)

def _tbl_by_scenario_and_k(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["k"].isin([10, 20]) & df_in["scenario"].isin(["gamma", "normal"])]
    pt = d.groupby(["method", "scenario", "k"])[value_col].sum().unstack(["scenario", "k"], fill_value=0)
    for sc in ["gamma","normal"]:
        for k in [10,20]:
            if (sc,k) not in pt.columns:
                pt[(sc,k)] = 0
    out = pd.DataFrame(index=pt.index)
    out["a_10"] = pt[("gamma",10)]
    out["a_20"] = pt[("gamma",20)]
    out["a_all"] = out[["a_10","a_20"]].sum(axis=1)
    out["b_10"] = pt[("normal",10)]
    out["b_20"] = pt[("normal",20)]
    out["b_all"] = out[["b_10","b_20"]].sum(axis=1)
    out.columns.name = None    # << Fix
    return _reindex_methods_left(out)


def _tbl_by_p(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["scenario"].isin(["gamma", "normal"]) & (df_in["k"] > 0) & df_in["p"].isin(P_LEVELS)]
    g = d.groupby(["method", "p"])[value_col].sum().unstack("p", fill_value=0)
    # Spalten sicherstellen und in gewünschter Reihenfolge
    for pp in P_LEVELS:
        if pp not in g.columns:
            g[pp] = 0
    g = g[P_LEVELS]
    # Optional: Spalten als "p=1000" usw. beschriften
    g.columns = [f"p={pp}" for pp in g.columns]
    return _reindex_methods_left(g)

def build_improvement_tables(df_src: pd.DataFrame, value_col: str, tag: str):
    """Erstellt und speichert alle gewünschten Tabellen für value_col ∈ {'Better','Better2'}."""
    t_scen = _tbl_by_scenario(df_src, value_col)
    t_k    = _tbl_by_k(df_src, value_col)
    t_sk   = _tbl_by_scenario_and_k(df_src, value_col)
    t_p    = _tbl_by_p(df_src, value_col)

    # Abspeichern als CSV
    t_scen.to_csv(os.path.join(PREP_DIR, f"Sim2_impr_by_scenario_{tag}.csv"), index=False)
    t_k.to_csv(   os.path.join(PREP_DIR, f"Sim2_impr_by_k_{tag}.csv"),         index=False)
    t_sk.to_csv(  os.path.join(PREP_DIR, f"Sim2_impr_by_scen_k_{tag}.csv"),    index=False)
    t_p.to_csv(   os.path.join(PREP_DIR, f"Sim2_impr_by_p_{tag}.csv"),         index=False)

    return {
        "by_scenario": t_scen,
        "by_k": t_k,
        "by_scenario_k": t_sk,
        "by_p": t_p,
    }

# --- Build & Save: mBIC (Better) und mBIC2 (Better2) ---
tabs_mBIC  = build_improvement_tables(df, "Better",  tag="mBIC")
tabs_mBIC2 = build_improvement_tables(df, "Better2", tag="mBIC2")
tabs_worse_mBIC  = build_improvement_tables(df, "Worse",  tag="mBIC_worse")
tabs_worse_mBIC2 = build_improvement_tables(df, "Worse2", tag="mBIC2_worse")
```

```python
from IPython.display import HTML, display
import pandas as pd

def _df_to_html(obj, max_rows=50, index=False):
    if obj is None or (hasattr(obj, "empty") and getattr(obj, "empty")):
        return "<em>(no data)</em>"
    if isinstance(obj, pd.DataFrame):
        return obj.head(max_rows).to_html(index=index)
    if hasattr(obj, "to_html"):
        try:    return obj.to_html()
        except TypeError:
            try: return obj.to_html(index=index)
            except Exception: pass
    return f"<pre>{repr(obj)}</pre>"

def _display_row(items, widths=None, max_height=420,
                 nowrap_first_col=True, table_width="100%", table_layout="fixed"):
    """
    items: [(Titel, DataFrame/Styler), ...]
    table_width: '100%' lässt die Tabelle den Card-Rahmen füllen
    table_layout: 'fixed' → gleichmäßigere Spalten, ellipsis greift
    """
    n = len(items)
    if n == 0:
        return
    if widths is None:
        widths_css = " ".join(["1fr"] * n)
    elif isinstance(widths, (list, tuple)):
        widths_css = " ".join(str(w) for w in widths)
    else:
        widths_css = str(widths)

    nowrap_css = "white-space: nowrap;" if nowrap_first_col else ""

    cards = []
    for title, df in items:
        cards.append(f"""
          <div class="card">
            <h3>{title}</h3>
            <div class="tablewrap">{_df_to_html(df)}</div>
          </div>
        """)

    html = f"""
    <style>
      .tblgrid {{
        display: grid;
        grid-template-columns: {widths_css};
        gap: 16px;
        align-items: start;
        margin: 12px 0;
      }}
      .tblgrid .card h3 {{
        margin: 0 0 8px 0;
        font-size: 16px;
        line-height: 1.2;
      }}
      .tblgrid .tablewrap {{
        max-height: {max_height}px;
        overflow: auto;              /* Scrollbar falls nötig */
        border: 1px solid #e0e0e0;
        border-radius: 6px;
      }}
      .tblgrid table {{
        border-collapse: collapse;
        font-size: 13px;
        width: {table_width};        /* <-- füllt den Rahmen */
        table-layout: {table_layout};/* 'fixed' = gleichmäßig, ellipsis funktioniert */
      }}
      .tblgrid th, .tblgrid td {{
        padding: 4px 8px;
        border: 1px solid #ddd;
        text-align: right;
        overflow: hidden;            /* für ellipsis */
        text-overflow: ellipsis;     /* … */
        white-space: nowrap;         /* Zellen bleiben einzeilig */
      }}
      .tblgrid th:first-child, .tblgrid td:first-child {{
        text-align: left;
        {nowrap_css}                 /* Methoden-Spalte einzeilig */
      }}
    </style>
    <div class="tblgrid">{''.join(cards)}</div>
    """
    display(HTML(html))
```

```python
# ===== Mittelwert-Tabellen-Builder (für Power, FP, FDR, Runtime) =====
SC_MAP   = {"gamma": "a", "normal": "b"}
K_LEVELS = [10, 20]
P_LEVELS = [1000, 2000, 5000, 10000]

def _reindex_methods_left(pt: pd.DataFrame) -> pd.DataFrame:
    pt = pt.reindex(method_order)
    pt.index.name = "method"
    pt.columns.name = None
    return pt.reset_index()

def _avg_by_scenario(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["scenario"].isin(["gamma","normal"]) & (df_in["k"] > 0)]
    g = d.groupby(["method","scenario"])[value_col].mean().unstack("scenario", fill_value=0.0)
    g = g.rename(columns=SC_MAP)
    for c in ["a","b"]:
        if c not in g.columns: g[c] = 0.0
    # "all" als Mittel über beide Szenarien (über alle Zeilen)
    all_mean = d.groupby("method")[value_col].mean()
    g["all"] = g.index.map(all_mean.to_dict()).astype(float)
    g = g[["a","b","all"]]
    return _reindex_methods_left(g)

def _avg_by_k(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["k"].isin(K_LEVELS) & df_in["scenario"].isin(["gamma","normal"])]
    g = d.groupby(["method","k"])[value_col].mean().unstack("k", fill_value=0.0)
    for k in K_LEVELS:
        if k not in g.columns: g[k] = 0.0
    # "all" als Mittel über beide k
    all_mean = d.groupby("method")[value_col].mean()
    g["all"] = g.index.map(all_mean.to_dict()).astype(float)
    g = g.rename(columns={10:"k=10", 20:"k=20"})
    g = g[["k=10","k=20","all"]]
    return _reindex_methods_left(g)

def _avg_by_p(df_in: pd.DataFrame, value_col: str) -> pd.DataFrame:
    d = df_in[df_in["scenario"].isin(["gamma","normal"]) & (df_in["k"] > 0) & df_in["p"].isin(P_LEVELS)]
    g = d.groupby(["method","p"])[value_col].mean().unstack("p", fill_value=0.0)
    for pp in P_LEVELS:
        if pp not in g.columns: g[pp] = 0.0
    g = g[P_LEVELS]
    g.columns = [f"p={pp}" for pp in g.columns]
    return _reindex_methods_left(g)

def build_avg_tables(df_src: pd.DataFrame, value_col: str, tag: str):
    t_scen = _avg_by_scenario(df_src, value_col)
    t_k    = _avg_by_k(df_src, value_col)
    t_p    = _avg_by_p(df_src, value_col)

    # speichern
    t_scen.to_csv(os.path.join(PREP_DIR, f"Sim2_avg_by_scenario_{tag}.csv"), index=False)
    t_k.to_csv(   os.path.join(PREP_DIR, f"Sim2_avg_by_k_{tag}.csv"),         index=False)
    t_p.to_csv(   os.path.join(PREP_DIR, f"Sim2_avg_by_p_{tag}.csv"),         index=False)

    return {"by_scenario": t_scen, "by_k": t_k, "by_p": t_p}
```

```python
def plot_facets_metric(df_src, value_col, title_main, out_png,
                       SCENARIOS=('gamma','normal'), K_LEVELS=(10,20),
                       SC_MAP={'gamma':'a','normal':'b'},
                       colors=None, linestyles=None):
    dsrc = df_src[(df_src['k'].isin(K_LEVELS)) & (df_src['scenario'].isin(SCENARIOS))].copy()

    # Methodenreihenfolge wie im Notebook (nur vorhandene)
    if 'method_order' in globals() and method_order:
        methods = [m for m in method_order if m in set(dsrc['method'])]
    else:
        methods = list(pd.unique(dsrc['method']))

    # x-Achse & Jitter
    p_ticks = sorted(dsrc['p'].unique())
    dx = np.diff(p_ticks).min() if len(p_ticks) > 1 else 1.0
    jitter_frac = 0.12
    offsets = (np.linspace(-0.5, 0.5, len(methods)) * dx * jitter_frac) if len(methods) > 1 else [0.0]
    offset_map = dict(zip(methods, offsets))

    def _series_for_panel(dsrc, k, scenario):
        d = dsrc[(dsrc['k'] == k) & (dsrc['scenario'] == scenario)]
        out = {}
        for m in methods:
            y = []
            for p in p_ticks:
                v = d.loc[(d['method'] == m) & (d['p'] == p), value_col]
                y.append(float(v.iloc[0]) if not v.empty and v.iloc[0] == v.iloc[0] else 0.0)
            x = [p + offset_map[m] for p in p_ticks]
            out[m] = (x, y)
        return out

    # Plot 2×2
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(11.5, 7.5), sharex=True, sharey=True)
    for ci, sc in enumerate(SCENARIOS):       # Spalten: a/b
        for ri, k in enumerate(K_LEVELS):     # Zeilen: 10/20
            ax = axes[ri, ci]
            series = _series_for_panel(dsrc, k, sc)
            for m in methods:
                x, y = series[m]
                ax.plot(
                    x, y,
                    label=m,
                    color=(colors or {}).get(m, None),
                    linestyle=(linestyles or {}).get(m, '-'),
                    linewidth=2,
                    marker='o', markersize=4, alpha=0.95, zorder=2
                )
            ax.set_title(SC_MAP.get(sc, sc), fontsize=12, pad=6)
            ax.grid(True, linestyle=':', linewidth=0.6, alpha=0.6, zorder=1)
            # k rechts an den Panels
            if ci == len(SCENARIOS) - 1:
                ax.text(1.04, 0.5, f"{k}", transform=ax.transAxes,
                        ha='left', va='center', fontsize=12, fontweight='normal')

    # Achsen & Ticks
    for ax in axes[-1, :]:
        ax.set_xticks(p_ticks)
        ax.set_xlabel("p", labelpad=6)
    for ax in axes[:, 0]:
        ax.set_ylabel("")

    # Titel & Legende
    fig.suptitle(title_main, fontsize=15, y=0.97)

    handles = [Line2D([0], [0],
                      color=(colors or {}).get(m, 'gray'),
                      linestyle=(linestyles or {}).get(m, '-'),
                      linewidth=2, marker='o', markersize=4, label=m)
               for m in methods]
    fig.legend(handles=handles, title="method", loc="center left",
               bbox_to_anchor=(0.985, 0.5), borderaxespad=0.4, frameon=True)

    # Layout enger + Platz rechts
    plt.subplots_adjust(left=0.07, right=0.92, top=0.86, bottom=0.10, wspace=0.15, hspace=0.18)

    # Speichern + zeigen
    os.makedirs(PREP_DIR, exist_ok=True)
    plt.savefig(out_png, dpi=150, bbox_inches='tight')
    print("Gespeichert:", out_png)
    plt.show()
```

```python
def plot_mbic_values_facets(df_long, value_col, title, subtitle,
                            p_levels=None, k_levels=(10, 20),
                            methods=None, exclude="deep2stage",
                            out_png=None):
    """
    df_long: df_mBIC oder df_mBIC2 im Long-Format (Spalten: variable, k, p, scenario, <value_col>)
    value_col: 'mBIC' oder 'mBIC2'
    methods: explizite Reihenfolge der Methoden (optional); sonst aus Daten
    exclude: Set/Liste von Methoden, die ausgeblendet werden sollen
    """
    d = df_long.copy()
    d = d.rename(columns={'variable': 'method'})
    d = d[d['k'].isin(k_levels)]
    d = d[d['scenario'].isin(['gamma','normal'])]  # a/b zusammen wie im Screenshot

    # p-Spalten
    if p_levels is None:
        p_levels = sorted(d['p'].unique())
    else:
        p_levels = [p for p in p_levels if p in set(d['p'])]

    # Methoden-Reihenfolge
    if methods is None:
        if 'method_order' in globals() and method_order:
            methods = [m for m in method_order if m in set(d['method'])]
        else:
            methods = list(d['method'].dropna().unique())
    if exclude:
        methods = [m for m in methods if m not in set(exclude)]

    # Figure/Axes
    nrows, ncols = len(k_levels), len(p_levels)
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(2.2*ncols*2.2, 2.2*nrows*3.0), sharey=False)
    if nrows == 1: axes = np.array([axes])
    if ncols == 1: axes = axes.reshape(nrows, 1)

    # Style (leichtes ggplot-Grau)
    face_grey = (0.95, 0.95, 0.95)

    # Jitter rund um die diskreten Methoden-Positionen
    x_pos = np.arange(len(methods))
    rng = np.random.default_rng(42)
    jitter = 0.12

    for r, k in enumerate(k_levels):
        for c, p in enumerate(p_levels):
            ax = axes[r, c]
            ax.set_facecolor(face_grey)
            ax.grid(True, linestyle=':', linewidth=0.6, alpha=0.7)

            sub = d[(d['k'] == k) & (d['p'] == p)].copy()

            # Scatter pro Methode (schwarz wie im Beispiel)
            scat_data = []
            for j, m in enumerate(methods):
                y = sub.loc[sub['method'] == m, value_col].dropna().to_numpy()
                if y.size:
                    xj = x_pos[j] + rng.normal(0, jitter, size=y.size)
                    ax.scatter(xj, y, s=12, color='black', alpha=0.8, edgecolor='none')
                    scat_data.append(y)
                else:
                    scat_data.append(np.array([]))

            # Boxplots (über alle Sim-werte je Methode)
            # nur Methoden mit Daten; Positionen entsprechend
            bp_vals = [y for y in scat_data]
            ax.boxplot(
                bp_vals,
                positions=x_pos,
                widths=0.6,
                whis=(5, 95),
                showfliers=False,
                patch_artist=True,
                boxprops=dict(facecolor='white', edgecolor='black', linewidth=1),
                medianprops=dict(color='black', linewidth=1.2),
                whiskerprops=dict(color='black', linewidth=0.8),
                capprops=dict(color='black', linewidth=0.8)
            )

            # 0-Linie (Referenz = stepwise_plain)
            ax.axhline(0, color='black', linewidth=1)

            # Achse formatieren
            ax.set_xticks(x_pos)
            ax.set_xticklabels([])
            ax.tick_params(axis='x', length=0)
            ax.set_xticks(x_pos)
            if r == nrows - 1:
                ax.set_xticklabels(methods, rotation=90)
            else:
                ax.set_xticklabels([])
                ax.tick_params(axis='x', length=0)
            if c == 0:
                ax.set_ylabel("mBIC" if value_col == "mBIC" else "mBIC2")

            # Spaltentitel "p = ..."
            if r == 0:
                ax.set_title(f"p = {p}", fontsize=11, pad=8)

            # Reihenlabel rechts wie Facet-Strip: "k = ..."
            if c == ncols - 1:
                ax.text(1.02, 0.5, f"k = {k}", transform=ax.transAxes,
                        ha='left', va='center', fontsize=11, rotation=-90)

    # Haupttitel + Untertitel
    fig.suptitle(title, fontsize=14, y=0.98)
    fig.text(0.08, 0.95, subtitle, fontsize=10)

    # Layout
    plt.subplots_adjust(left=0.08, right=0.96, top=0.90, bottom=0.18, wspace=0.20, hspace=0.22)

    # Speichern
    if out_png:
        os.makedirs(os.path.dirname(out_png), exist_ok=True)
        plt.savefig(out_png, dpi=150, bbox_inches="tight")
        print("Gespeichert:", out_png)

    plt.show()

# ------------------ Aufrufe ------------------
EXCLUDE = {"deep2stage","GSDAR"} 

# mBIC (Differenzen zur Baseline aus df_mBIC)
plot_mbic_values_facets(
    df_mBIC, value_col="mBIC",
    title="mBIC values", subtitle="stepwise_plain serves as reference",
    p_levels=[1000, 2000, 5000, 10000, 50000],  # nimm die, die du hast
    k_levels=(10, 20),
    methods=[m for m in method_order if m in set(df_mBIC['variable'])],  # konsistente Reihenfolge
    exclude=EXCLUDE,  # z.B. {"deep2stage", "deep2stage_plus"}
    out_png=os.path.join(PREP_DIR, "Sim2_mBIC_values_facets.png")
)

# mBIC2 (Differenzen aus df_mBIC2)
plot_mbic_values_facets(
    df_mBIC2, value_col="mBIC2",
    title="mBIC2 values", subtitle="stepwise_plain serves as reference",
    p_levels=[1000, 2000, 5000, 10000, 50000],
    k_levels=(10, 20),
    methods=[m for m in method_order if m in set(df_mBIC2['variable'])],
    exclude=EXCLUDE,
    out_png=os.path.join(PREP_DIR, "Sim2_mBIC2_values_facets.png")
)
```

```python
import os
import numpy as np
import pandas as pd
from IPython.display import display, HTML

# ------------------ Parameter ------------------
SC_LABEL = {"gamma": "Gamma", "normal": "Normal"}   # Spaltenbeschriftungen
SC_ORDER = ["gamma", "normal"]                      # Spalten-Reihenfolge
K_LEVELS = [10, 20]                                 # Modellgrößen (falls andere: anpassen)
DROP_METHODS = {"stepwise_plain"}                   # Baseline auslassen (optional)

# ------------------ Helper ------------------
def _method_list(d: pd.DataFrame, method_order=None):
    ms = list(d["method"].dropna().unique())
    if method_order:
        ms = [m for m in method_order if m in set(ms)]
    return ms

def _mk_block(dsrc: pd.DataFrame, value_col: str, block_title: str,
              scenarios, k_levels, methods):
    """
    Baut einen Spaltenblock (Better/Worse) mit Column-MultiIndex:
      (block_title, Distribution, k)
    Werte = Summe über p.
    """
    if value_col not in dsrc.columns:
        # Leerer Block mit korrektem Spalten-Layout
        cols = [(block_title, SC_LABEL[s], k) for s in scenarios for k in k_levels]
        empty = pd.DataFrame(index=methods, columns=pd.MultiIndex.from_tuples(cols))
        return empty.fillna(0).astype(int)

    g = (dsrc.groupby(["method", "scenario", "k"], as_index=False)[value_col]
              .sum(min_count=1))
    # Pivot → Spalten (scenario, k)
    pv = (g.pivot(index="method", columns=["scenario", "k"], values=value_col)
            .reindex(index=methods))
    # gewünschte Spaltenreihenfolge und vollständige Matrix
    cols = []
    for s in scenarios:
        for k in k_levels:
            if (s, k) not in pv.columns:
                pv[(s, k)] = 0
            cols.append((s, k))
    pv = pv[cols].fillna(0).astype(int)

    # MultiIndex-Spalten: (Better/Worse, Distribution, k)
    new_cols = pd.MultiIndex.from_tuples(
        [(block_title, SC_LABEL[s], int(k)) for (s, k) in pv.columns],
        names=["Criterion", "Distribution", "Model size"]
    )
    pv.columns = new_cols
    pv.index.name = "method"
    return pv

def build_impr_worse_table(df: pd.DataFrame, method_order=None,
                           drop_methods=None, k_levels=None, scenarios=None):
    """
    Gibt eine Tabelle mit zwei Zeilenblöcken zurück:
      Indexlevel 0: 'mBIC' / 'mBIC2'
      Indexlevel 1: Methoden
    Spalten: Better/Worse × Distribution × k
    """
    if k_levels is None:
        k_levels = K_LEVELS
    if scenarios is None:
        scenarios = SC_ORDER

    # Vorbereiten
    d = df.copy()
    d = d.rename(columns={"variable": "method"})
    if drop_methods:
        d = d[~d["method"].isin(set(drop_methods))]
    d = d[d["k"].isin(k_levels) & d["scenario"].isin(scenarios)]

    # Methoden in gewünschter Reihenfolge
    methods = _method_list(d, method_order=method_order)

    # --- mBIC: Better/Worse ---
    d_m = d.copy()
    block_better_m = _mk_block(d_m, "Better", "Better criterion", scenarios, k_levels, methods)
    block_worse_m  = _mk_block(d_m, "Worse",  "Worse criterion",  scenarios, k_levels, methods)
    tab_m = pd.concat([block_better_m, block_worse_m], axis=1)
    tab_m.index = pd.MultiIndex.from_product([["mBIC"], tab_m.index], names=["Metric", "method"])

    # --- mBIC2: Better2/Worse2 ---
    d_m2 = d.copy()
    block_better_m2 = _mk_block(d_m2, "Better2", "Better criterion", scenarios, k_levels, methods)
    block_worse_m2  = _mk_block(d_m2, "Worse2",  "Worse criterion",  scenarios, k_levels, methods)
    tab_m2 = pd.concat([block_better_m2, block_worse_m2], axis=1)
    tab_m2.index = pd.MultiIndex.from_product([["mBIC2"], tab_m2.index], names=["Metric", "method"])

    # Zusammenfügen
    final = pd.concat([tab_m, tab_m2], axis=0)

    # Spalten sortieren: Better vor Worse; innerhalb: Gamma:10,20 → Normal:10,20
    def _sort_cols(mi):
        crits = ["Better criterion", "Worse criterion"]
        dists = [SC_LABEL[s] for s in scenarios]
        ks    = [int(k) for k in k_levels]
        order = [(c, dist, k) for c in crits for dist in dists for k in ks]
        cols_existing = [t for t in order if t in set(mi)]
        # ggf. vergessene Spalten anhängen
        extras = [t for t in mi if t not in cols_existing]
        return cols_existing + extras

    final = final.reindex(columns=_sort_cols(final.columns))
    return final

# ------------------ Bauen, speichern, anzeigen ------------------
# Optional: Methoden ausschließen (Baseline etc.)
drop_methods = DROP_METHODS

summary_tbl = build_impr_worse_table(
    df, method_order=method_order, drop_methods=drop_methods,
    k_levels=[10, 20], scenarios=["gamma", "normal"]
)

os.makedirs(PREP_DIR, exist_ok=True)
out_csv = os.path.join(PREP_DIR, "Sim2_impr_worse_by_dist_k_mBIC_mBIC2.csv")
summary_tbl.to_csv(out_csv)
print("Gespeichert:", out_csv)

# hübsch anzeigen
display(HTML("<h3 style='margin: 12px 0 6px 0;'>Number of improvements / worsenings by distribution and model size</h3>"))
try:
    show(summary_tbl)  # falls du show() im Notebook nutzt
except Exception:
    display(summary_tbl)
```

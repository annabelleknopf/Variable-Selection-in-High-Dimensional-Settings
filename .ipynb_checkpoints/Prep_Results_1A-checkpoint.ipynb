# Prep Results — Simulation 1A


```python
import os, re, pickle, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import matplotlib as mpl
from pathlib import Path

def _display_tbl(name, tbl):
    if tbl is None:
        return
    try:
        if hasattr(tbl, "empty") and tbl.empty:
            display(HTML(f"<h3 style='margin: 1.25em 0 0.5em 0'>{name}</h3><p><em>(no data)</em></p>"))
            return
    except Exception:
        pass
    display(HTML(f"<h3 style='margin: 1.25em 0 0.5em 0'>{name}</h3>"))
    try:
        show(tbl)   # pretty renderer if available
    except Exception:
        display(tbl)  # fallback
```

```python
mpl.rcParams.update({
    "figure.dpi": 150,
    "savefig.dpi": 300,
    "axes.grid": True,
    "grid.alpha": 0.3,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "boxplot.patchartist": False,   # unfilled boxes (no colors)
    "lines.linewidth": 1.5,
    "axes.labelsize": 12,
    "axes.titlesize": 13,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
})

# Toggle for saving figures (PNG + PDF)
SAVE_FIGS = False
FIGDIR = "figures"
os.makedirs(FIGDIR, exist_ok=True)
```

```python
# Optional explicit override
RESULTS_FILE = None  

def _discover_results():
    patterns = [
        re.compile(r"(?i)sim1a\.k_20\.rho_0\.5\.(pkl|pickle)$"),
        re.compile(r"(?i)sim1a\.(pkl|pickle)$"),
    ]
    roots = [
        ".",
        "CheckResults1", "Results1",
        "CheckResults1A", "Results1A",
        str(Path.cwd()),
        str(Path.cwd().parent),
        str(Path.home()),
    ]
    candidates = []
    for root in roots:
        if not os.path.exists(root):
            continue
        for dp, dn, fnames in os.walk(root):
            for fn in fnames:
                for pat in patterns:
                    if pat.search(fn):
                        full = os.path.join(dp, fn)
                        try:
                            mtime = os.path.getmtime(full)
                        except Exception:
                            mtime = 0
                        candidates.append((mtime, full))
                        break
    candidates.sort(reverse=True)  # newest first
    return [c[1] for c in candidates]

if RESULTS_FILE and os.path.exists(RESULTS_FILE):
    path = RESULTS_FILE
    found = [path]
else:
    found = _discover_results()
    path = found[0] if found else None

if not path:
    raise FileNotFoundError(
        "Could not find a Sim1A results pickle.\n"
        "Tips:\n"
        "  • Ensure the simulation saved e.g. 'Sim1a.k_20.rho_0.5.pkl'.\n"
        "  • Set RESULTS_FILE to the exact path.\n"
        "  • Or place the file under one of: ./, CheckResults1/, Results1/."
    )

print("Using results file:", path)

with open(path, "rb") as f:
    data = pickle.load(f)

# Arrays (sim_nr, nr_methods)
mBIC_results  = np.asarray(data.get("mBIC_results"))
mBIC2_results = np.asarray(data.get("mBIC2_results"))
runtime       = np.asarray(data.get("runtime"))
method_names  = list(data.get("method_names"))
k = data.get("k", 20)
rho = data.get("rho", 0.5)

# Optional arrays to derive FP/FDR/Power (if present)
mBIC_FP  = data.get("mBIC_FP", None)
mBIC2_FP = data.get("mBIC2_FP", None)
mBIC_TP  = data.get("mBIC_TP", None)
mBIC2_TP = data.get("mBIC2_TP", None)
mBIC_FP  = np.asarray(mBIC_FP)  if mBIC_FP  is not None else None
mBIC2_FP = np.asarray(mBIC2_FP) if mBIC2_FP is not None else None
mBIC_TP  = np.asarray(mBIC_TP)  if mBIC_TP  is not None else None
mBIC2_TP = np.asarray(mBIC2_TP) if mBIC2_TP is not None else None

method_order = method_names[:]
sim_nr, nr_methods = mBIC_results.shape
print(f"sims={sim_nr}, methods={nr_methods}, k={k}, rho={rho}")
```

```python
def _safe_div(a, b):
    return np.divide(a, b, out=np.full_like(a, np.nan, dtype=float), where=(b!=0))

FDR2  = _safe_div(mBIC2_FP.astype(float), (mBIC2_FP + mBIC2_TP).astype(float)) if (mBIC2_FP is not None and mBIC2_TP is not None) else None
Power = (mBIC_TP.astype(float)  / float(k)) if (mBIC_TP  is not None) else None
Power2= (mBIC2_TP.astype(float) / float(k)) if (mBIC2_TP is not None) else None
```

```python
def to_long(metric_array, metric_name):
    rows = []
    if metric_array is None:
        return pd.DataFrame(columns=["sim","method","metric","value"])
    for j, m in enumerate(method_order):
        col = metric_array[:, j]
        for i, val in enumerate(col):
            rows.append((i, m, metric_name, val))
    return pd.DataFrame(rows, columns=["sim","method","metric","value"])

df_mBIC    = to_long(mBIC_results,  "mBIC")
df_mBIC2   = to_long(mBIC2_results, "mBIC2")
df_runtime = to_long(runtime,       "Runtime")
df_FP_mBIC = to_long(mBIC_FP,       "FP")
df_FDR2    = to_long(FDR2,          "FDR2")
df_Power   = to_long(Power,         "Power")
df_Power2  = to_long(Power2,        "Power2")

for d in (df_mBIC, df_mBIC2, df_runtime, df_FP_mBIC, df_FDR2, df_Power, df_Power2):
    if not d.empty:
        d["method"] = pd.Categorical(d["method"], categories=method_order, ordered=True)


def simple_boxplot(
    df_metric,
    methods=None,            # Reihenfolge der Methoden; default: unique in df_metric oder method_order
    title="",                # Plot-Titel
    ylabel=None,             # Y-Achsentitel; default = title
    showfliers=True,         # Ausreißer ein/aus
    notch=False,             # Notch aus = keine „Sanduhr“
    showmeans=True,          # Mittelwert-Linie anzeigen
    label_rotation=90,       # Drehung der Methodennamen
    ylim=None,               # z.B. (0, 10) oder None
    zero_line=False,         # 0-Linie (praktisch für Δ-Plots)
    save_base=None           # Basis-Dateiname ohne Endung, speichert PNG+PDF wenn gesetzt
):
    if methods is None:
        methods = list(df_metric["method"].dropna().unique())
    data = [df_metric.loc[df_metric["method"] == m, "value"].dropna().values for m in methods]

    fig, ax = plt.subplots(figsize=(10, 5))

    # Schwarz-weiß Eigenschaften (keine Füllung)
    bw_props = dict(
        boxprops     = dict(color="black", linewidth=1.0),
        capprops     = dict(color="black", linewidth=1.0),
        whiskerprops = dict(color="black", linewidth=1.0),
        medianprops  = dict(color="black", linewidth=1.2),
        meanprops    = dict(color="black"),
        flierprops   = dict(marker="o", markersize=3, markerfacecolor="none",
                            markeredgecolor="black", linewidth=0.6),
    )

    ax.boxplot(
        data,
        labels=methods,      # Matplotlib <3.9: 'labels' (>=3.9 ginge auch 'tick_labels')
        showfliers=showfliers,
        notch=notch,
        showmeans=showmeans, meanline=True,
        patch_artist=False,  # keine Füllung -> wirklich S/W
        **bw_props
    )

    ax.tick_params(axis="x", labelrotation=label_rotation)
    if ylabel is None:
        ylabel = title or "value"
    ax.set_ylabel(ylabel if ylabel is not None else (title or "value"),
                  labelpad=20)  # <- mehr Abstand
    #ax.set_title(title)
    if ylim is not None:
        ax.set_ylim(*ylim)
    if zero_line:
        ax.axhline(0, linestyle="--", linewidth=1)

    ax.grid(True, axis="y", alpha=0.3)
    plt.tight_layout()

    if save_base:
        plt.savefig(f"{save_base}.png", bbox_inches="tight")
        plt.savefig(f"{save_base}.pdf", bbox_inches="tight")

    plt.show()
    return fig, ax


# Outliers an, S/W, Labels 90°
simple_boxplot(df_mBIC,    methods=method_order, title="mBIC",    ylabel="mBIC",    showfliers=True, notch=False, label_rotation=90)
simple_boxplot(df_mBIC2,   methods=method_order, title="mBIC2",   ylabel="mBIC2",   showfliers=True, notch=False, label_rotation=90)
simple_boxplot(df_runtime, methods=method_order, title="Runtime", ylabel="seconds", showfliers=True, notch=False, label_rotation=90)
```

```python

def summarise_metric(df_metric):
    if df_metric is None or df_metric.empty:
        return pd.DataFrame()
    def q1(x): return np.nanpercentile(x, 25)
    def q3(x): return np.nanpercentile(x, 75)
    grp = (df_metric
           .groupby("method", observed=True)["value"]
           .agg(mean=np.nanmean,
                std=np.nanstd,
                median=np.nanmedian,
                q1=q1, q3=q3,
                n=lambda x: x.notna().sum())
          )
    return grp.reset_index()

tbl_mBIC    = summarise_metric(df_mBIC)
tbl_mBIC2   = summarise_metric(df_mBIC2)
tbl_runtime = summarise_metric(df_runtime)
tbl_FP      = summarise_metric(df_FP_mBIC)
tbl_FDR2    = summarise_metric(df_FDR2)
tbl_Power   = summarise_metric(df_Power)
tbl_Power2  = summarise_metric(df_Power2)

_display_tbl("Summary statistics — mBIC (k=20, rho=0.5)",    tbl_mBIC.round(4))
_display_tbl("Summary statistics — mBIC2 (k=20, rho=0.5)",   tbl_mBIC2.round(4))
_display_tbl("Summary statistics — Runtime (k=20, rho=0.5)", tbl_runtime.round(4))

_display_tbl("Summary statistics — FP (mBIC)",   tbl_FP.round(4))
_display_tbl("Summary statistics — FDR (mBIC2)", tbl_FDR2.round(4))
_display_tbl("Summary statistics — Power (mBIC)",  tbl_Power.round(4))
_display_tbl("Summary statistics — Power (mBIC2)", tbl_Power2.round(4))
```
